**Question 8:** Write a Python program to convert Celsius to Fahrenheit.

```python
# Your code here
#following code is to demonstrate temperature conversion from celcius to fahrenheit and vice versa
Temperature = input("Enter the temperature you want to onvert (e.g., 45c or 102f): ")
if Temperature[-1] in ['c','C']:
    celsius  = float(Temperature[:-1])
    fahrenheit = (celsius * 9/5) + 32
    print(f"{celsius}°C is equal to {fahrenheit}°F")
elif Temperature[-1] in ['f','F']:
    fahrenheit = float(Temperature[:-1])
    celsius = (fahrenheit - 32) * 5/9
    print(f"{fahrenheit}°F is equal to {celsius}°C")
else:
    print("Invalid input format. Please enter the temperature followed by 'c' or 'f'.")
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
# Leap year logic:\n
# - A year is a leap year if it is divisible by 4,\n
# - EXCEPT years divisible by 100 are NOT leap years,\n
# - UNLESS the year is also divisible by 400 (then it is a leap year).\n
# This implements the standard Gregorian calendar rule.

def is_leap_year(year):
    """Return True if `year` is a leap year, otherwise False."""
    # Years divisible by 400 are leap years.
    if year % 400 == 0:
        return True
    # Years divisible by 100 (but not 400) are NOT leap years.
    if year % 100 == 0:
        return False
    # Years divisible by 4 (but not 100) are leap years.
    if year % 4 == 0:
        return True
    return False

# Example usage:
try:
    year = int(input("Enter a year: "))
    if is_leap_year(year):
        print(f"{year} is a leap year.")
    else:
        print(f"{year} is not a leap year.")
except ValueError:
    print("Please enter a valid integer year.")
```
